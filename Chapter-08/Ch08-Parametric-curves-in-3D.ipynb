**Example: Coil curve**

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,10*np.pi,500)
r = 1
x = r * np.cos(u)
y = r * np.sin(u)
z = 2 * u
fig = px.line_3d(x=x,y=y,z=z)
fig.update_traces(line_color='darkgreen',line_width=5)
fig.show()
```

**Example: Flower shape curve**

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,2*np.pi,500)
r = 2 + np.sin(5*u)
x = r * np.cos(u)
y = r * np.sin(u)
z = np.abs(x) * np.abs(y)
fig = px.line_3d(x=x,y=y,z=z)
fig.update_traces(line_color='darksalmon',line_width=6)
fig.show()
```

**Twister curve**

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,10*np.pi,500)
r = u**2
x = r * np.cos(u)
y = r * np.sin(u)
z = 2 * u
fig = px.line_3d(x=x,y=y,z=z)
fig.update_traces(line_color='blueviolet',line_width=10)
fig.show()
```
