**Modeling a four-elements meshgrid surface**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(1,2,2)
v = np.linspace(3,4,2)
U = np.array([[u[0],u[1]], [u[0],u[1]]])
print('U: ', U)
V = np.array([[v[0],v[0]], [v[1],v[1]]])
print('V: ', V)
X = 2*U + V
Y = U + V**2
plt.fill([X[0,0],X[1,0], X[1,1],X[0,1]], [Y[0,0], Y[1,0],Y[1,1],Y[0,1]])
plt.show()
```

**Modeling an eight-elements meshgrid surface**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(1,4,4)
v = np.linspace(3,4,2)
U = np.array([[u[0],u[1],u[2],u[3]],[u[0],u[1],u[2],u[3]]])
print('U: ', U)
V = np.array([[v[0],v[0],v[0],v[0]],[v[1],v[1],v[1],v[1]]])
print('V: ', V)
X = U**2 + V
Y = 3*U + V**2
for col in range(len(u)-1):
  plt.fill([X[0,col],X[1,col],X[1,col+1],X[0,col+1]],[Y[0,col],Y[1,col],Y[1,col+1],Y[0,col+1]],
           color='blue',alpha=.5)
plt.show()
```

**Manual definition of a large meshgrid size**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(1,4,4)
v = np.linspace(3,7,5)
U = np.array([[u[0],u[1],u[2],u[3]], [u[0],u[1],u[2],u[3]], [u[0],u[1],u[2],u[3]], [u[0],u[1],u[2],u[3]], [u[0],u[1],u[2],u[3]]])
print('U: ', U)
V = np.array([[v[0],v[0],v[0],v[0]], [v[1],v[1],v[1],v[1]], [v[2],v[2],v[2],v[2]], [v[3],v[3],v[3],v[3]], [v[4],v[4],v[4],v[4]]])
print('V: ', V)
```

**Using a function to define a large meshgrid size**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(1,4,4)
v = np.linspace(3,7,5)
U,V = np.meshgrid(u,v)
print('U: ', U)
print('V: ', V)
```

**Modeling a large meshgrid size surface**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(1,4,4)
v = np.linspace(3,7,5)
U,V = np.meshgrid(u,v)
print('U: ', U)
print('V: ', V)
X = U**2 * V
Y = U * V**3
for row in range(len(v)-1):
  for col in range(len(u)-1):
    plt.fill([X[row,col],X[row+1,col],X[row+1,col+1],X[row,col+1]],
             [Y[row,col],Y[row+1,col],Y[row+1,col+1],Y[row,col+1]], color='green',alpha=.4)
plt.show()
```

**Example: Disc with a hole**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(0,2*np.pi,50)
v = np.linspace(1,5,10)
U,V = np.meshgrid(u,v)
r = 2 * V
X = r * np.sin(U)
Y = r * np.cos(U)
for row in range(len(v)-1):
  for col in range(len(u)-1):
    plt.fill([X[row,col],X[row+1,col],X[row+1,col+1],X[row,col+1]],
             [Y[row,col],Y[row+1,col],Y[row+1,col+1],Y[row,col+1]],color='chocolate',alpha=.5)
plt.show()
```

**Example: Star with seven picks**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(0,2*np.pi,80)
v = np.linspace(1,10,10)
U,V = np.meshgrid(u,v)
r = V + 2*np.sin(7*U) + 5
X = r * np.cos(U)
Y = r * np.sin(U)
for row in range(len(v)-1):
  for col in range(len(u)-1):
    plt.fill([X[row,col],X[row+1,col],X[row+1,col+1],X[row,col+1]],
             [Y[row,col],Y[row+1,col],Y[row+1,col+1],Y[row,col+1]],color='firebrick',alpha=.7)
plt.show()
```

**The Reuleaux triangle**

```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(0,2*np.pi,60)
v = np.linspace(0,1,10)
U,V = np.meshgrid(u,v)
r = V
X = r * (-8*(np.sin(U)**3-1)*np.cos(U))
Y = r * (-2*np.cos(2*U)-np.cos(4*U)+8*np.sin(U))
for row in range(len(v)-1):
  for col in range(len(u)-1):
    plt.fill([X[row,col],X[row+1,col],X[row+1,col+1],X[row,col+1]],
             [Y[row,col],Y[row+1,col],Y[row+1,col+1],Y[row,col+1]],color='seagreen',alpha=.7)
plt.show()
```
