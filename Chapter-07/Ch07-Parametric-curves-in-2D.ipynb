**Explicit notation of curves in 2D**

```python
import numpy as np
import plotly.express as px
u = np.linspace(1,5,200)
x = u
y = np.log10(x) * np.sin(x**2) + 1
fig = px.line(x=x,y=y)
fig.update_traces(line_color='black',line_width=4)
fig.show()
```

```python
import numpy as np
import plotly.express as px
u = np.linspace(-np.pi/4,np.pi/4,50)
x = u
y = -np.pi/5 + np.sqrt(1-x**2)
fig = px.line(x=x,y=y)
fig.update_traces(line_color='blue',line_width=4)
```

```python
import numpy as np
import plotly.express as px
u = np.linspace(-np.pi,5*np.pi,100)
y = u
x = y + 3*np.sin(y)**3
fig = px.line(x=x,y=y)
fig.update_traces(line_color='red',line_width=4)
```

**Implicit notation of curves in 2D**

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,2*np.pi,100)
r = 1
x = r * np.cos(u)
y = r * np.sin(u)
fig = px.line(x=x,y=y)
fig.update_traces(line_color='purple',line_width=5)
fig.update_yaxes(scaleanchor="x",scaleratio=1)
```

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,16*np.pi,1000)
r = u**2
x = r * np.cos(u)
y = r * np.sin(u)
fig = px.line(x=x,y=y)
fig.update_traces(line_color='black',line_width=4)
fig.update_yaxes(scaleanchor="x",scaleratio=1)
```

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,2*np.pi,100)
r = 2 + np.sin(5*u)
x = r * np.cos(u)
y = r * np.sin(u)
fig = px.line(x=x,y=y)
fig.update_traces(line_color='fuchsia',line_width=5)
fig.update_yaxes(scaleanchor="x",scaleratio=1)
```

```python
import numpy as np
import plotly.express as px
u = np.linspace(0,2*np.pi,5000)
A = 1
a = 1
B = 0.6
b = -2
deltha = np.pi/20
x = A * np.sin(a*u + deltha)
y = B * np.sin(b*u)
fig = px.line(x=x,y=y)
fig.update_traces(line_color='blue',line_width=30)
fig.update_yaxes(scaleanchor="x",scaleratio=1)
```
