**plot function**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6,10,-5,-4])
y = np.array([0,1, 5, 7, 3])
plt.plot(x,y,color='black',linewidth=3)
plt.plot(x,y,'o',color='red',markersize=7)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Raster Graph using plot function')
plt.show()
```

**fill function**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6,10,-5,-4])
y = np.array([0,1, 5, 7, 3])
plt.fill(x,y,color='red',alpha=0.4,linewidth=3)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Raster Graph using fill function')
plt.show()
```
